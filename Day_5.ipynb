```python
# Task is to merge data of 2 files : one of type csv and another xlsx
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv('data_x.csv')
df1
```

```python
# Drop rows from a dataframe
df1 = df1.drop(df1.index[20:])
df1
```

```python
df2 = pd.read_excel('data_x.xlsx')
df2
```

```python
# To concatenate 2 dataframe their structure should be same...
frames = [df1, df2] # store the names of the dataframe in a list named frames
result = pd.concat(frames)
result
```

```python
result.loc[0]
```

```python
# after concatenating the indexes need to be reset
result = result.reset_index()
result
```

```python
# drop the additional index column
result = result.drop(['index'], axis=1)
result
```

```python
result.to_excel('dump_x.xlsx', index=False)
```

```python
# Joins
```

```python
df_Emp = pd.read_excel('Emp_1.xlsx')
df_Yos = pd.read_excel('Emp_Yos.xlsx')
```

```python
df_Emp
```

```python
df_Yos
```

```python
'''
Types of joins :
inner ==> Intersection (the common values)
outer ==> Union (all the records)
left ==> Left join (all records of the left dataset)
right ==> Right join (all records of the right dataset)
'''
```

```python
pd.merge(df_Emp, df_Yos, how='inner')
```

```python
pd.merge(df_Emp, df_Yos, how='outer')
```

```python
pd.merge(df_Emp, df_Yos, how='left')
```

```python
pd.merge(df_Emp, df_Yos, how='right')
```

```python
pd.merge(df_Emp, df_Yos, how='inner', on=['EmpID'])
```

```python
# line plot of Speed (km/hr) v/s Calories burnt
speed = [10, 20, 30, 40]
cal_sunday = [100, 175, 225, 300]
cal_monday = [120, 195, 245, 320]
cal_tuesday = [90, 170, 235, 305]
plt.plot(speed , cal_sunday , color = 'hotpink' , label = 'Sunday', marker = 'o')
plt.plot(speed , cal_monday , color = 'lightgreen' , label = 'Monday', marker = '*')
plt.plot(speed , cal_tuesday , color = 'red' , label = 'Tuesday', marker = '^')
plt.title('Speed v/s Calories burnt')
plt.xlabel('Speed (km/hr)')
plt.ylabel('Calories Burnt')
plt.legend()
plt.grid()
plt.show()
```
