```python
# dataframe.at funtion is used to update a certain value in the dataframe
# df[row , col]
```

```python
import datetime
now = datetime.datetime.now()
date_time = now.strftime('%d-%m-%Y')
date_time
```

```python
df.at[27, 'Date'] = date_time
df
```

```python
# Finding duplicate rows
```

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.info()
```

```python
# Correcting the Duration column
```

```python
for row in df.itertuples():
    if df.at[row.Index, 'Duration'] == 450:
        df.at[row.Index, 'Duration'] = 45
    elif df.at[row.Index, 'Duration'] == 600:
        df.at[row.Index, 'Duration'] = 60
```
