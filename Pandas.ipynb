```python
# Series create , manipulate , querry , delete
import pandas as pd

ls = [0,1,2,3,4,5]
s1 = pd.Series(ls) # remember to write the "S" in upper case while writting Series .
print(s1)

print()
order = [1,2,3,4,5,6]
s2 = pd.Series(ls , index = order)
print(s2)

print()
# Index can aslo be the string or character values :
import numpy as np
n = np.random.randn(5) # creating a random NDarray
index = ["a","b","c","d","e"]
s3 = pd.Series(n , index = index)
print(s3)

print()
# Create series from dictionary .
d = {"a":1,"b":2,"c":3,"d":4,"e":5}
s4 = pd.Series(d)
print(s4)

# we can modify the index of series :
print()
print(s1)
print()
s1.index = ['A','B','C','D','E','F']
print(s1)

print()
# we can also do slicing here :
s1[ : 3]
s1[ : -2]
s1[2 : ]
s1[-2 :]

print()
s5 = s1.append(s3)
print(s5)
print()
print(s5.drop("e"))
```

```python
 # Series Operations :
import pandas as pd

arr  = [0,1,2,3,4,5,7]
arr2 = [6,7,8,9,5]

s6 = pd.Series(arr2)
print(s6)
print()
s7 = pd.Series(arr)
print(s7)
print()
print(s6.add(s7))
print()
print(s6.sub(s7))
print()
print(s6.mul(s7))
print()
print(s6.div(s7))
print()
print("Median",s6.median())
print("Max",s6.max())
print("Min",s6.min())

```

```python
# Create DataFrame :
import pandas as pd
import numpy as np

dates = pd.date_range("today" , periods = 6) # Defining time sequence as index
print(dates)
print()
num_arr = np.random.randn(6,4) # Import numpy random array .
print(num_arr)
print()
columns = ['A','B','C','D'] # Use the table as the column name
print(columns)
print()
df1 = pd.DataFrame(num_arr , index = dates , columns = columns)
print(df1)

print()
# creating a dataframe with dictionary array
data = {'animal': ['cat','cat','snake','dog','dog','cat','snake','cat','dog','dog'],
        'age': [2.5,3,0.5,np.nan,5,2,4.5,np.nan,7,3],
        'visits':[1,3,2,3,2,3,1,1,2,1],
        'priority': ['yes','yes','no','yes','no','no','no','yes','no','no']}

labels = ['a','b','c','d','e','f','g','h','i','j']

df2 = pd.DataFrame(data, index = labels)
print(df2)
print()
# we can also see the datatypes of array :
print(df2.dtypes)
print()

# if we want to see the top values of the table then we will use head .
# for ex :
print(df2.head()) # By default it will show the starting 5 values .
# if we want to see only specified values then we can mention it in our parameters .
print()
print(df2.head(3))
print()
# if we want to take this 3 values into another dataframe we can do that too.
df3 = df2.head(3)
print(df3)

# another one is tail() method which will return the last values .
print()
print(df2.tail()) # by default it will also give the data of 5 values from the last .
# we can do same operations on tail() as we have performed it on head() method .

print()
# we can also see the index , columns and values of dataframe separately .
# for ex :
print(df2.index) # we cannot give parameters to index because it's not a callable object .
print()
print(df2.columns) # same as above note for index .
print()
print(df2.values) # same as above note for index .

print()
""" * Seeing the statitical data of dataframe """
print(df2.describe())
print()
# trying to manipulate the dataframe .
print(df2.T) # the rows are converted into columns and columns are converted into rows .
print()
print(df2.sort_values(by = 'age')) # we can also sort the values according to the our convinience .

print()
# we can also perform the slicing on dataframe .
print(df2[1:3])

print()
# query dataframe by tag .
print(df2[['age','visits']]) # we'll get only age and visits to see .
print()
print(df2.iloc[1:3]) # Query rows 2,3 (iloc = integer location)
print()
# we also have copy funciton to use here .
df3 = df2.copy()
print(df3)
print()
# we can aslo see the null values present .
print(df3.isnull())
print()
# we can also change the values within a dataframe .
df3.loc['f','age']=1.5
print(df3)
print()
df3.loc["f", ['age', 'visits']] = 2, 4 # we can do this also or we can change the values separately .
print(df3)
print()
# we can also calclulate the mean
print(df3.mean())
print()
# we can also do this by specifying the column name .
print(df3[["age"]].mean())
print()
# we can aslo do the sum() of visits .
print(df3['visits'].sum())
print()
print(df3['visits'].min())
print()
print(df3['visits'].max())
print()
# we can aslo do the sum of everything .
print(df3.sum()) # for strings it will concatenate everything and for integers it wil give the total .
print()
string = pd.Series(["A","C",'D','Aaa','BaCa',np.nan,'CBA','cow','owl'])
string.str.upper() # we can also do it lower() .
```

```python
""" Operations for DataFrame missing values : """
import pandas as pd


df4 = df2.copy()
meanAge = df4['age'].mean()
# in the age their are NaN values which are present , to fill those values we can use 'fillna()'
df4["age"].fillna(meanAge)
```

```python
# we can also do :
df5 = df2.copy()
print(df5.fillna(4))
print()
# we can also drop the row which has missing data .
df6 = df2.copy()
print(df6.dropna(how="any"))

```
